```python
import torch
```

```python
# Создайте случайный FloatTensor размера 3x4x5

A = torch.randn(size=(3,4,5),dtype=torch.float64)
A
```

```python
# Выведите его форму (shape)
A.shape
```

```python
# Приведите его к форме 6 X 10
A = A.reshape((6,10))
A
```

```python
# Умножьте его на вектор [1, 4, 2, 2, 1] поэлементно
B = torch.tensor(data=[1,4,2,2,1],dtype=torch.float64)

A = A.reshape((3,4,5))
C = A * B
print(C)
```

```python
# Умножьте тензор матрично на себя, чтобы результат был размерности 6x6
AA1 = A.reshape((6,10))
print(AA1,end='\n\n')
AA2 = AA1.transpose(0,1)
print(AA2,end='\n\n')
AA = torch.matmul(AA1, AA2)
print(AA)
```

```python
# Посчитайте производную функции y = x**3 + z - 75t в точке (1, 0.5, 2)
x = torch.tensor(1, dtype=torch.float64, requires_grad=True)
z = torch.tensor(0.5, dtype=torch.float64, requires_grad=True)
t = torch.tensor(2, dtype=torch.float64, requires_grad=True)

y = x**3 + z - 75*t
print('Значение функции: {}'.format(y))
y.backward()
print('Значения производных: x_d={}, z_d={}, t_d={}'.format(x.grad,z.grad,t.grad))
```
